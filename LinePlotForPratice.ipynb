Practice Question:

Using the Seaborn library, load the "tips" dataset and create a line plot to visualize the trend of total bill amounts across different days of the week.
 Customize the plot by:

Setting an appropriate title.

Changing the line style and color.

Adding markers to show individual data points.

Improving readability with labels for the x-axis and y-axis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y1=sns.load_dataset("tips")
y1
av_bill=y1.groupby("day")["total_bill"].mean().reset_index()
av_bill
```

```python
sns.lineplot(x="day",y="total_bill",data=av_bill,style="total_bill",markers=True)

plt.title("Average Bill by Day")
plt.grid(color='red')
plt.ylabel("Average Bill")
plt.show()
```
