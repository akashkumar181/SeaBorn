```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("healthexp")
var
```

```python
sns.displot(var["Spending_USD"],bins=[0,2000,4000,6000,8000,10000,12000],kde=True,color='g',rug=True)
plt.show()
```

Using Log Scale


```python
sns.displot(var["Spending_USD"],log_scale=True,kde=True,color='g',rug=True)
plt.show()
```
