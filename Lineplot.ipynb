```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y1=sns.load_dataset("penguins")
y1
y1.head(10)
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y1,hue="sex",style="sex",palette="Accent",markers=["o",">"])
plt.grid(color='r')
plt.show()
```
